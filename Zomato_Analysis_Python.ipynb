```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel('/content/BangaloreZomatoDataAnalysis clean.xlsx')

# Check data
print("Total Restaurants:", len(df))
print("Columns:", df.columns.tolist())
```

```python
# Chart 1 — Top 10 Areas by Restaurant Count
plt.figure(figsize=(12,6))
area_count = df['Area'].value_counts().head(10)
sns.barplot(x=area_count.values, y=area_count.index, palette='viridis')
plt.title('Top 10 Areas by Number of Restaurants', fontsize=15)
plt.xlabel('Number of Restaurants')
plt.ylabel('Area')
plt.tight_layout()
plt.savefig('chart1_area.png')
plt.show()
```

```python
# Chart 2 — Delivery Rating Distribution
plt.figure(figsize=(10,6))
df['Delivery Ratings'] = pd.to_numeric(df['Delivery Ratings'], errors='coerce')
sns.histplot(df['Delivery Ratings'].dropna(), bins=20, color='coral', kde=True)
plt.title('Distribution of Delivery Ratings', fontsize=15)
plt.xlabel('Delivery Rating')
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.savefig('chart2_ratings.png')
plt.show()
```

```python
# Chart 3 — Veg vs Non-Veg
plt.figure(figsize=(8,8))
veg_counts = df['isVegOnly'].value_counts()
labels = ['Non-Veg', 'Veg Only']
colors = ['#ff9999','#90EE90']
plt.pie(veg_counts.values, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title('Veg vs Non-Veg Restaurants in Bangalore', fontsize=15)
plt.tight_layout()
plt.savefig('chart3_veg.png')
plt.show()
```

```python
# Chart 4 — Average Cost by Area (Top 10)
plt.figure(figsize=(12,6))
cost_by_area = df.groupby('Area')['AverageCost'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=cost_by_area.values, y=cost_by_area.index, palette='magma')
plt.title('Top 10 Most Expensive Areas in Bangalore', fontsize=15)
plt.xlabel('Average Cost for Two (₹)')
plt.ylabel('Area')
plt.tight_layout()
plt.savefig('chart4_cost.png')
plt.show()
```

```python
# Chart 5 — Top 10 Areas by Delivery Rating
plt.figure(figsize=(12,6))
df['Delivery Ratings'] = pd.to_numeric(df['Delivery Ratings'], errors='coerce')
top_rated = df.groupby('Area')['Delivery Ratings'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_rated.values, y=top_rated.index, palette='coolwarm')
plt.title('Top 10 Areas by Average Delivery Rating', fontsize=15)
plt.xlabel('Average Delivery Rating')
plt.ylabel('Area')
plt.xlim(3.5, 4.5)
plt.tight_layout()
plt.savefig('chart5_ratings_area.png')
plt.show()
```
